```python
import random 

def consultar_tabla_periodica():
    print("\n" + " TABLA PERIÓDICA COMPLETA ".center(50, '='))
    print("\nSi no entiendes algún término, puedes ir al diccionario para que te lo explique.")
    
    while True:
        print("\nOpciones:")
        print("1. Buscar elemento por nombre")
        print("2. Ver diccionario de términos")
        print("3. Elemento al azar") 
        print("4. Salir")
        
        opcion = input("Seleccione (1-4): ")
        
        if opcion == '1':
            nombre = input("\nNombre del elemento (sin tilde): ").lower().strip()
            if nombre in elementos:
                mostrar_elemento(nombre)
            else:
                print("Elemento no encontrado. Intente nuevamente.")
        elif opcion == '2':
            mostrar_diccionario()
        elif opcion == '3':  
            elemento_aleatorio = random.choice(list(elementos.keys()))
            print(f"\n¡Elemento aleatorio seleccionado!: {elemento_aleatorio.capitalize()}")
            mostrar_elemento(elemento_aleatorio)
        elif opcion == '4':
            print("\n¡Hasta prooonto!")
            break
        else:
            print("Opción no válida. Intente nuevamente.")

# Llamada correcta a la función
consultar_tabla_periodica() 
```
