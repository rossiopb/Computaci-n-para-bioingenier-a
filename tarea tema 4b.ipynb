```python
import csv
from statistics import mean

# 1. Leer el archivo CSV
ruta_archivo = r'C:\Users\josue\OneDrive\Desktop\COMISION REVISORA ACTUAL\Computación para bioingeniería\Tareas Python\prevention_of_mother_to_child_transmission_by_country_clean.csv'

try:
    with open(ruta_archivo, mode='r', encoding='utf-8') as archivo:
        lector_csv = csv.reader(archivo)
        
        # Convertir el lector a lista para poder trabajar con él
        datos = list(lector_csv)
        
        # 2. Contar filas y columnas
        num_filas = len(datos)
        if num_filas > 0:
            num_columnas = len(datos[0])
        else:
            num_columnas = 0
        
        print(f"\n1. El archivo tiene {num_filas} filas y {num_columnas} columnas")
        
        # Opción para incluir/excluir encabezados
        incluir_encabezados = input("\n2. ¿Incluir encabezados en los resultados? (s/n): ").lower() == 's'
        inicio_datos = 0 if incluir_encabezados else 1
        
        # 3. Imprimir primeros 5 registros 
        print("\n3. Primeros 5 registros:")
        contador = 0
        for i in range(inicio_datos, num_filas):
            if contador < 5:
                print(datos[i])
                contador += 1
            else:
                break

        # 4. Verificar valores únicos en una columna específica
        if num_filas > 0:
            try:
                columna = int(input(f"\n4. Ingrese el índice de columna para ver valores únicos (0-{num_columnas-1}): "))
                if 0 <= columna < num_columnas:
                    valores_columna = [fila[columna] for fila in datos[inicio_datos:] if columna < len(fila)]
                    valores_unicos = set(valores_columna)
                    print(f"\nValores únicos en columna {columna}: {len(valores_unicos)}")
                    print(f"Muestra de valores únicos: {list(valores_unicos)[:5]}...")
                else:
                    print("Índice de columna inválido")
            except ValueError:
                print("Debe ingresar un número válido")

        # 5. Calcular media de columna numérica
        if num_filas > 0:
            try:
                columna_num = int(input(f"\n5. Ingrese el índice de columna numérica para calcular media (0-{num_columnas-1}): "))
                if 0 <= columna_num < num_columnas:
                    valores_numericos = []
                    for fila in datos[inicio_datos:]:
                        if columna_num < len(fila):
                            try:
                                valor = float(fila[columna_num])
                                valores_numericos.append(valor)
                            except ValueError:
                                continue
                    
                    if valores_numericos:
                        media = mean(valores_numericos)
                        print(f"\nMedia de columna {columna_num}: {media:.2f}")
                    else:
                        print("No se encontraron valores numéricos en esta columna")
                else:
                    print("Índice de columna inválido")
            except ValueError:
                print("Debe ingresar un número válido")

        # 6. Encontrar máximo y mínimo en columna numérica
        if num_filas > 0 and 'valores_numericos' in locals() and valores_numericos:
            maximo = max(valores_numericos)
            minimo = min(valores_numericos)
            print(f"\n6. Valores extremos en columna {columna_num}:")
            print(f"Máximo: {maximo:.2f}")
            print(f"Mínimo: {minimo:.2f}")

        # 7. Función sorpresa: Detección de valores atípicos
        if num_filas > 1 and 'valores_numericos' in locals() and len(valores_numericos) > 1:
            print("\n7.  ¡Sorpresa! Detección de valores atípicos:")
            media_val = mean(valores_numericos)
            desviacion = (sum((x - media_val)**2 for x in valores_numericos)/len(valores_numericos))**0.5
            
            if desviacion > 0:
                umbral_superior = media_val + 2*desviacion
                umbral_inferior = media_val - 2*desviacion
                
                atipicos = [x for x in valores_numericos if x > umbral_superior or x < umbral_inferior]
                
                print(f"Media: {media_val:.2f}")
                print(f"Desviación estándar: {desviacion:.2f}")
                print(f"Valores atípicos (fuera de {umbral_inferior:.2f} - {umbral_superior:.2f}):")
                print(f"Cantidad: {len(atipicos)}")
                print(f"Valores: {sorted(atipicos)}")
            else:
                print("No se detectaron valores atípicos (desviación estándar cero)")

except FileNotFoundError:
    print("Error: No se encontró el archivo en la ruta especificada")
except Exception as e:
    print(f"Ocurrió un error: {e}")
        
               
           
```
