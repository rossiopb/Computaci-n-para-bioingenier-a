```python



from docx import Document
from docx.shared import Pt, RGBColor, Inches
from docx.enum.text import WD_PARAGRAPH_ALIGNMENT
from docx.oxml import OxmlElement
from docx.oxml.ns import qn

def create_contract_template():
    # Crear un nuevo documento
    doc = Document()
    
    # 1. Crear tabla para logo + nombre (1 fila, 2 columnas)
    table = doc.add_table(rows=1, cols=2)
    table.autofit = False  # Desactivar autofit para control manual

    # Configurar celdas
    cells = table.rows[0].cells

    # Celda 1: Insertar logo (ajusta la ruta y tamaño)
    logo_cell = cells[0]
    logo_cell.width = Inches(1.5)  # Ancho para el logo
    logo_paragraph = logo_cell.paragraphs[0]
    logo_paragraph.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER
    logo_run = logo_paragraph.add_run()
    logo_run.add_picture("C:\Tareas001\logo.jpeg", width=Inches(1.0))  # Ajusta el tamaño

    # Celda 2: Nombre de la compañía (Times New Roman, negrita, centrado)
    name_cell = cells[1]
    name_cell.width = Inches(4.0)
    name_paragraph = name_cell.paragraphs[0]
    name_paragraph.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER
    name_run = name_paragraph.add_run("Jueguitoys")
    name_run.font.name = 'Times New Roman'
    name_run.font.size = Pt(36)
    name_run.font.bold = True

    # 2. INFORMACIÓN DE CONTACTO (Arial)
    # Dirección
    direccion = doc.add_paragraph("[Torre Centrum – Piso 7, Oficina 704, Cañada Strongest, a media cuadra de la Plaza del Estudiante.], [La Paz, 0000]")
    direccion.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER
    direccion.runs[0].font.size = Pt(10)
    direccion.runs[0].font.name = 'Arial'
    
    # Teléfono y Fax
    contacto = doc.add_paragraph("[Celular: 78878174][Fax:123-123-123456]")
    contacto.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER
    contacto.runs[0].font.size = Pt(10)
    contacto.runs[0].font.name = 'Arial'
    
    # Email
    email = doc.add_paragraph("[jueguitoys@gmail.com]")
    email.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER
    email.runs[0].font.size = Pt(10)
    email.runs[0].font.name = 'Arial'
    
    # Espacio en blanco
    doc.add_paragraph("")
    
    # 3. TÍTULO DEL CONTRATO (Arial)
    title = doc.add_paragraph("Contrato de Compraventa")
    title.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER
    title.runs[0].font.size = Pt(37)  
    title.runs[0].font.bold = True
    title.runs[0].font.underline = True
    title.runs[0].font.name = 'Arial'
    
    # Espacio en blanco
    doc.add_paragraph("")

    
    acuerdo = doc.add_paragraph()
    acuerdo_run = acuerdo.add_run("ACUERDO")
    acuerdo_run.bold = True
    acuerdo_run.font.name = 'Calibri'

    # 2. Texto normal  continuado
    acuerdo.add_run("celebrado el día 28 de marzo de 2025, entre Luis Miguel Carrasco Sanchez (Primera Parte), y Sebastian Santander Huanca (Segunda Parte)")
    acuerdo.runs[1].font.name = 'Calibri' 

    # 4. Testimonio
    testimonio = doc.add_paragraph("TESTIMONIO:")
    testimonio.add_run("celebrado el día 28 de marzo de 2025, entre Luis Miguel Carrasco Sanchez (Primera Parte), y Sebastian Santander Huanca (Segunda Parte)")
    testimonio.runs.font.size = Pt(37)  
    testimonio.font.bold = True
    testimonio.font.name = 'Calibri'

        
    # Sección I
    section1 = doc.add_paragraph("I.")
    section1.add_run("\n\nSaid party of the first part covenants and agrees that it shall:")
    
    for run in section1.runs:
        run.font.name = 'Times New Roman'
        run.font.size = Pt(12)
    
    
    # Texto introductorio
    intro = doc.add_paragraph("THIS AGREEMENT made this")
    intro.add_run("\n\nday of\n\n20\n\nby and\n\nbetween\n\n(First Party) and\n\n(Second Party).")
    intro.runs[0].font.size = Pt(12)
    
    # Espacio en blanco
    doc.add_paragraph("")
    
    # Texto WITNESSETH
    witness = doc.add_paragraph("WITNESSETH: That in consideration of the mutual covenants and agreements to be kept and")
    witness.add_run("\n\nperformed on the part of said parties hereto, respectively as herein stated:")
    witness.runs[0].font.size = Pt(12)
    
    # Espacio en blanco
    doc.add_paragraph("")
    
    # Sección I
    section1 = doc.add_paragraph("I.")
    section1.add_run("\n\nSaid party of the first part covenants and agrees that it shall:")
    section1.runs[0].font.size = Pt(12)
    section1.runs[1].font.size = Pt(12)
    
    # Guardar el documento
    doc.save("Contract_Template.docx")
    print("Documento creado exitosamente: Contract_Template.docx")

# Ejecutar la función
    create_contract_template()
```
